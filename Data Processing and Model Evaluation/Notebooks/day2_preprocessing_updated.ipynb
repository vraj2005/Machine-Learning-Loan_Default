# Day 2 – Data Preprocessing & Clean CSV Creation
This updated notebook includes each preprocessing step in separate code cells.

## Step 1: Load Dataset

```
import pandas as pd

df = pd.read_csv("../Dataset/Loan_default.csv")
df.head()
```

## Step 2: Drop LoanID Column

```
df.drop(columns=['LoanID'], inplace=True)
df.head()
```

## Step 3: Convert Yes/No to 0/1

```
binary_cols = ['HasMortgage', 'HasDependents', 'HasCoSigner']

for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

df.head()
```

## Step 4: One-Hot Encode Categorical Columns

```
categorical_cols = ['Education', 'EmploymentType', 'MaritalStatus', 'LoanPurpose']

df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
df.head()
```

## Step 5: Save Clean CSV

```
df.to_csv("../data/clean.csv", index=False)
print("clean.csv created successfully!")
```
